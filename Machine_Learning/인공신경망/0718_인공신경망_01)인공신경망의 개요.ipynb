```python
from sklearn.datasets import make_moons

# noise 표준편차
X, y = make_moons(n_samples=100, noise=0.25, random_state=3)
```

```python
from sklearn.model_selection import train_test_split
# stratify=y : 학습용 데이터와 검증용 데이터에서 각각의 타겟값들이 동일한 비율로 나타나도록함
X_train, X_test, y_train, y_test = train_test_split(
    X,y,test_size=0.2,stratify=y,random_state=0
)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.neural_network import MLPClassifier
model = MLPClassifier(random_state=0, max_iter=5000).fit(X_train, y_train)
# hidden_layer_sized=(100,), 은닉층에서 사용할 유닛슈의 기본값 : 100
# 데이터가 매우 적은 상태이므로 100개를 쓰는 것보다 줄이는 것이 좋다.
# 은닉유닛을 줄이면 모델이 복잡도는 낮아진다.

print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```

```python
from matplotlib import pyplot as plt
mglearn.plots.plot_2d_separator(model,X_train,fill=True,alpha=0.3)
mglearn.discrete_scatter(X_train[:,0], X_train[:,1], y_train)
plt.xlabel('class 0')
plt.xlabel('class 1')
```

```python
# 은닉유닛을 10개로 줄이고 데스트
model = MLPClassifier(random_state=0, hidden_layer_sizes=[10], max_iter=1000)
model.fit(X_train, y_train)

print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```

```python
mglearn.plots.plot_2d_separator(model, X_train, fill=True, alpha=0.3)
mglearn.discrete_scatter(X_train[:,0], X_train[:,1], y_train)
plt.xlabel('class 0')
plt.ylabel('class 1')
```

```python
# 10개의 유닛으로 된 두개의 은닉층 [10, 10]
model = MLPClassifier(random_state=0, hidden_layer_sizes=[10, 10], max_iter=500)
model.fit(X_train, y_train)

print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```

```python
mglearn.plots.plot_2d_separator(model, X_train, fill=True, alpha=.3)
mglearn.discrete_scatter(X_train[:,1], y_train)

plt.xlabel('class 0')
plt.ylabel('class 1')
```

```python
# 100개의 유닛으로 된 세 개의 은닉층 [100, 100, 100]
model = MLPClassifier(random_state=0, hidden_layer_sizes=[100, 100, 100], max_iter=500)
model.fit(X_train, y_train)

print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```

```python
# 100개의 유닛으로 된 세 개의 은닉층 [100, 100, 100]
model = MLPClassifier(random_state=0, hidden_layer_sizes=[100, 100, 100], max_iter=5000)
model.fit(X_train, y_train)

print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```

```python
mglearn.plots.plot_2d_separator(model, X_train, fill=True, alpha=.3)
mglearn.discrete_scatter(X_train[:,0], X_train[:,1], y_train)
plt.xlabel('class 0')
plt.xlabel('class 1')
```
